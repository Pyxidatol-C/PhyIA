```python
import cv2
from extract import *
import matplotlib.pyplot as plt
```

```python
plt.rcParams["figure.figsize"] = (20,3)
```

```python
im = cv2.imread("raw-data/slits/100.png")
m, k = -0.046130952380952384, 186
```

```python
plt.imshow(im[:, :, 2])
plt.axis('off')
plt.show()
```

```python
def dy_to_pts(im, m, k):
    return (0, k), (im.shape[1] - 1, int(round(m * (im.shape[1] - 1) + k)))
```

```python
p1, p2 = dy_to_pts(im, m, k)
cv2.line(im, p1, p2, (255, 0, 0), 3, cv2.LINE_AA)

im = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)

plt.axis('off')
plt.imshow(im)
plt.savefig('processed-data/axis.png', dpi=256)
plt.show()
```

```python
plt.axis('off')
plt.imshow(np.array([pattern(im, (m, k))] * 500))
plt.show()
```

```python
patt = np.array(pattern(im, (m, k)))
edge = patt <= 75

plt.axis('off')
plt.plot(edge, 'x')
plt.show()
```
